### Viết chương trình Python sử dụng một trong 2 loại vòng lặp for hoặc while để tính tống các số nguyên từ 1 đến n. (n <= 1000)

```python
n = int(input())
s = 0
for i in range(1,n+1):
    s += i                          
print(s)    
```

### VIết chương trình nhập vào một số tới khi được một số nguyên dương thì thôi. Kiểm tra số này có phải là số nguyên tố hay không?

```python
n = int(input())
s = 0
for i in range(1,n+1):
    if n % i == 0:
        s += 1
if s == 2:
    print('True')
else:
    print('False')    
```

### Viết chương trình tính giai thừa của một số nguyên n cho trước (n!=1×2×…×n) với 0 < n < 100.

```python
n = int(input())
s = 1
for i in range(1,n+1):
    s *= i
print(s)    
```

### Viết chương trình nhập vào một số nguyên n và in ra màn hình số chữ số (trừ dấu) của số đó. Không dùng kiểu dữ liệu string.

```python
n = int(input())
s = 0
while n > 0:
    if n % 10>= 0:
        n = n // 10
        s += 1
print(s)         
```

### Viết chương trình đếm số lượng ước số chẵn của số nguyên dương n với n < 10^6


```python
n = int(input())
s = 0
for i in range(1,n):
    if n % i == 0 and i % 2 == 0:
        s += 1
print(s)        
```

### Cho hai số m và n là hai số nguyên dương nhập từ bàn phím. Viết chương trình tìm ước chung lớn nhất của hai số m và n.


```python
a, b = map(int,input().split())
s = 1
for i in range(1, min(a,b)):
    if a % i == 0 and b % i == 0:
        s = i
print(s)        
```

### chân gà chó

```python

def nt(n):
    s = 0
    if n < 2:
        return False
    for i in range(1, n + 1):
        if n % i == 0:
            s += 1
    if s == 2:
        return True
    else:
        return False
if __name__ == '__main__':
    a = int(input())
    if nt(a):
      print('ok')
    else:
      print('no')    
    
```
