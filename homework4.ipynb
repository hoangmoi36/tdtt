### W4A18 Nhập vào số nguyên dương a và b, in toàn bộ ước chung của a và b

```python
a, b = map(int,input().split())
for i in range (1, min(a,b)):
    if a % i == 0 and b % i == 0:
        print(i, end =' ')
```

### W4A21 Nhập vào số nguyên dương n, tính tổng các chữ số của n

```python
n = int(input())
s = 0
while n > 0:
    if n% 10 > 0:
        s+= n % 10
    n = n // 10
print(s)
```

### W4A22 Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ

```python
c = 0
l = 0
n = int(input())
while n > 0:
    if n % 10 > 0:
        if (n % 10) % 2 == 0:
            c += 1
        else:
            l += 1
    n = n // 10
print(c,l)
```

W4A23 Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n

```python
n = int(input())
s = 0
for i in range(1, n):
    for k in range(1, i+1):
        if n == s:
            break
        s += k
if s == n:
    print('ok')
else:
    print('ko')
```

Nhập vào A, tìm n nhỏ nhất sao cho 
W4A24 1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A

```python
A = int(input())

S = 0
n = 0

while S <= A:
    n += 1
    S += 1 / n

print(n)
```

Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1. 
W4A25 Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập

```python
s = []
while True:
    n = int(input())
    if n == -1:
        break
    s.append(n)
print(max(s), min(s))
```

Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt 
W4A26 quá A

```python
n = int(input())

a, b = 1, 1

while b <= n:
    a, b = b, a + b

print(a)
```

Cho một chuỗi chỉ gồm các ký tự chữ cái thường và chữ cái hoa, hãy đổi ký tự hoa thành 
W4A29 thường và thường thành hoa

```python
n = input()
s = ""

for i in n:
    if i.islower():
        s += i.upper()
    else:
        s += i.lower()

print(s)

```

### Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in W4A31 thường, bao nhiêu ký tự số
