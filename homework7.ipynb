Cho một mảng số nguyên không âm. Gọi mảng đó là "mảng đẹp" nếu 
mỗi phần tử x trong mảng xuất hiện đúng x lần. 
Kiểm tra xem mảng có phải là mảng đẹp hay không. Nếu không hãy xóa 
đi y số (y là nhỏ nhất có thể) để mảng còn lại thành mảng đẹp

```python
arr = list(map(int, input().split()))
arrd = {}
for i in arr:
    if i in arrd:
        arrd[i] += 1
    else:
        arrd[i] = 1
y = 0
check = True
for j in arrd:
    if arrd[j] != j:
        check = False
        if arrd[j] > j:
            y += arrd[j] - j
        else:
            y += arrd[j]
if check:
    print("YES")
else:
    print("NO", y)
```

Tìm phần tử xuất hiện nhiều hơn n/2 lần trong mảng (nếu có) với độ 
phức tạp thời gian là O(n), độ phức tạp bộ nhớ là O(1), - Boyer–Moore. 
Nếu không có phần tử nào thỏa mãn thì không in gì

```python
arr = list(map(int, input().split()))
arrd = {}
n = len(arr)
for i in arr:
    if i in arrd:
        arrd[i] += 1
    else:
        arrd[i] = 1
for j in arrd:
    if arrd[j] > n // 2:
        print(j)
        break
```

Cho N là ngày, T là tháng năm 2020, Tính xem đó là ngày thứ bao nhiêu 
của năm đó nếu đó là ngảy hợp lệ

```python
def tinhngay(n, t):
    match t:
        case 1|3|5|7|8|10|12:
            maxday = 31
        case 4|6|9|11:
            maxday = 30
        case 2:
            maxday = 29
        case _:
            maxday = -1
    if n < 1 or n > maxday or maxday == -1:
        return "INVALID"
    tong = n
    if t == 1:
        return tong
    for i in range(1, t):
        match i:
            case 1|3|5|7|8|10|12:
                tong += 31
            case 4|6|9|11:
                tong += 30
            case 2:
                tong += 29
    return tong
n, t = map(int, input().split())
print(tinhngay(n, t))
```

Đếm số lượng tên trong xâu kí tự. Tên được định nghĩa là xâu có độ dài

Tìm số có đúng k chữ số lớn nhất có thể trong một số có n chữ số (coi 
đầu vào là một số rất lớn - str). Không được hoán đổi vị trí các chữ s

```python
def amoi(n,k):

    sl_xoa = len(n)- k
    tich_luy = []

    for i in n:
        while  tich_luy and sl_xoa > 0 and tich_luy[-1]:
            tich_luy.pop()
            sl_xoa -= 1
        tich_luy.append(i)
    if sl_xoa > 0:
        tich_luy = tich_luy[:-sl_xoa]
    return "".join(tich_luy)

n = list(map(int,input().split()))
k = int(input())
amoi(n,k)
```

Cho một dãy ngoặc ()(){}{}[][]..., kiểm tra xem dãy ngoặc đã cho có hợp lệ
hay không

```python
s = input()
arr = {")":"(", "]":"[", "}":"{", ">":"<"}
stack = []
for i in s:
    if i in arr:
        if len(stack) == 0 or arr[i] != stack[-1]:
            print("NO")
            break
        stack.pop()
    else:
        stack.append(i)
print("YES" if len(stack) == 0 else "NO")
```
