### W5A3 Viết hàm kiểm tra một số có phải là số nguyên tố hay không?

```python
def snt(n):
    s = 0
    for i in range(1, n + 1):
        if n % i == 0:
            s += 1
    if s == 2:
        return True
    else:
        return False
a = int(input())
snt(a)
```

### Khoảng cách Hamming giữa hai số nguyên là số lượng vị trí khác nhau giữa hai dãy bits tương ứng của chúng. Ví dụ: Khoảng cách Hamming giữa 1 (1) và 4 (100) là 2. Cho hai số nguyên x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này. Khoảng cách Hamming là cái tên được đặt theo tên của Richard Hamming, người giới thiệu lý thuyết này trong tài liệu có tính cơ sở của ông về mã phát hiện lỗi và sửa lỗi (error-detecting and error-correcting codes). Nó được sử dụng trong kỹ thuật viễn thông để tính số lượng các bit trong một từ nhị phân (binary word) bị đổi ngược, như một hình thức để ước tính số lỗi xảy ra trong quá trình truyền thông, và vì thế, đôi khi, nó còn được gọi là khoảng cách tín hiệu (signal distance). Việc phân tích trọng số Hamming của các bit còn được sử dụng trong một số ngành, bao gồm lý thuyết tin học, lý thuyết mã hóa, và mật mã học. Tuy vậy, khi so sánh các dãy ký tự có chiều dài khác nhau, hay các dãy ký tự có xu hướng không chỉ bị thay thế không thôi, mà còn bị ảnh hưởng bởi dữ liệu bị lồng thêm vào, hoặc bị xóa đi, phương pháp đo lường phức tạp hơn, như khoảng cách Levenshtein (Levenshtein distance) là một phương pháp có tác dụng và thích hợp hơn.

```python
def hm(a, b):
    s = 0
    for i in range(len(a)):
        if a[i] == b[i]:
            s += 1
    return s
a = input()
b = input()
hm(a, b)
```

### Viết hàm trả về tổng các chữ số của số nguyên dương n


```python
def snd(n):
    s = 0
    while n > 0:
        if (n % 10) > 0:
            s += n % 10
        n = n // 10
    return s
snd(745713461)
```
