Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số 
nguyên x

```python
def ktra(n):
    s = 0
    tong = 0
    while n > 0:
        s += 1
        n = n // 10
    while n > 0:
        a = (n%10)**s
        tong += a
        n = n //10
    if tong == n:
        return True
    else:
        return False
x = int(input())
ktra(x)        
```

### VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên.

```python
def sdx(l):
    n = int(l)
    n1 = n
    rev = 0
    while n > 0:
        rev = rev * 10 + n % 10
        n //= 10
    if n1 == rev:
        return True
    else:
        return False

a = input()
print(sdx(a))
```

### W5A19 VIết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên.

### W5A27 Viết hàm trả về số lượng chữ số khác 0 trong n số nguyên nhập vào từ bàn phím (một list)

```python
def dem(lst):
    dem = 0
    for num in lst:
        for ch in str(abs(num)):
            if ch != '0':
                dem += 1
    return dem
```

W5A28 Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x1,y1), C(x3,y3)

```python
import math
def goc(x1,y1,x2,y2,x3,y3):
    xa = x1 - x2
    ya = y1 - y2
    xc = x3 - x2
    yc = y3 - y2
    tvh = xa * xc + ya * yc
    tdai = (( xa** 2 + ya ** 2)** 0.5)*( (xc ** 2 + yc ** 2)** 0.5)
    cosx = tvh / tdai
    goc_rad = math.acos(cosx)        
    goc_deg = math.degrees(goc_rad)  
    return goc_deg
x1,y1,x2,y2,x3,y3 = map(int,input().split())
goc(x1,y1,x2,y2,x3,y3)
```
