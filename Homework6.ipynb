Nhâp vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma 
W6A16 trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải

```python
n, m = map(int, input().split())

mt = []

for i in range(n):
    row = []
    for j in range(m):
        a = int(input())
        row.append(a)
    mt.append(row)

for i in range(n):
    for j in range(m):
        print(f"{mt[i][j]:>4}", end="")
    print()
```

Nhâp vào mảng n hàng, m cột số nguyên và số nguyên k (n, m, k nhập trên 
cùng một dòng, cách nhau dấu cách). Tính và in ra tổng các phần tử trên cột 
W6A18 k (hàng k)

```python
n, m, k = map(int, input().split())

matran = []

for i in range(n):
    row = []
    for j in range(m):
        a = int(input())
        row.append(a)
    matran.append(row)

tonghangk = 0
for x in matran[k-1]:
    tonghangk += x

tongcotk = 0
for i in range(n):
    for j in range(m):
        if j == k - 1:
            tongcotk += (matran[i][j])

print(tongcotk, tonghangk)
```

Viết hàm đưa vào 1 list số nguyên, tìm và trả về vị trí có giá trị lớn nhất 
W6A19 trong list

```python
def so_lon(a):
    max_num = 0
    for i in range(len(a)):
        if a[i]> max_num:
            max_num = a[i]
    return max_num
a = list(map(int,input().split()))
so_lon(a)
```

Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và 
trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về 
W6A20 -1

```python
def kiem_tra(a,k):
    vi_tri = -1
    for i in range(len(a)):
        if a[i] == k:
            vi_tri = i
            return vi_tri
s = list(map(int,input().split()))
h = int(input()) 
kiem_tra(s,h)          
```

Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ 
W6A21 chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện

```python
s = input().split()

d = {}
for i in range(len(s)):
    key = s[i]
    value = i
    if key in d:
        d[key].append(value)
    else:
        d[key] = [value]
for a in d:
    d[a] = tuple(d[a])
print(d)
```

VIết hàm xóa đi các số <x có trong mảng

```python
def delete(a,x):
    for i in a[:]:
        if i < x:
            a.remove(i)
    return a
s = list(map(int, input().split()))
x = int(input())
print(delete(s, x))
```

Tính tổng các phần tử cực đại địa phương trong một danh sách, không tính 2 
W6A26 phần tử đầu và cuối

```python
a = list(map(int,input().split()))
sum = 0
for i in range(1, len(a)-1):
    if  a[i-1] < a[i] and a[i] > a[i+1]:
        sum += (a[i])
print(sum)

```

Cho một dãy số nguyên, Tìm và in ra vị trí các số < x (x là giá trị cho trước). 
W6A27 nếu không có in ra -1

```python
s = list(map(int,input().split()))
x = int(input())
so_nho = []
for i in  range(len(s)):
    if s[i]< x:

        so_nho.append(s[i])
if not so_nho:
    print('-1')
else:
    print(so_nho)

```

Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm 
W6A28 thực sự, hay vô danh (không tăng, không giảm)

```python
a = list(map(int,input().split()))
tang = 1
giam = 1
for i in range(len(a)):
    if a[i] > a[i-1]:
        tang += 1
    elif a[i]< a[i-1]:
        giam += 1
if tang == len(a):
    print('tăng')
elif giam == len(a):
    print('giảm')   
else:
    print('vo danh')   
```
