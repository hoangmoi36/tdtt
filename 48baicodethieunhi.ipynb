Yêu cầu. Cước phí Taxi Thông Minh.
Viết một hàm duy nhất smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính và trả về total (tổng tiền phải trả, 
kiểu số nguyên, đơn vị VND). 
Các quy tắc cần cài đặt: 
1. Giá mở cửa (0–0.8 km): 
- Xe 4 chỗ: 11,000 VND 
- Xe 7 chỗ: 12,000 VND 
2. Giá theo km sau 0.8 km: 
- Xe 4 chỗ: từ 0.8→20 km ơnh 13,000 VND/km; trên 20 km tính 10,500 VND/km 
- Xe 7 chỗ: từ 0.8→20 km ơnh 14,000 VND/km; trên 20 km tính 12,000 VND/km 
3. Phí chờ: 5 phút đầu miễn phí; sau đó, mỗi 5 phút (làm tròn lên theo từng block 5 phút) tính 2,000 VND. 
4. Phụ thu ban đêm: nếu start_hour thuộc [22,23,0,1,2,3,4,5], cộng +20% trên tiền tạm tính (mở cửa + quãng đường + 
chờ). 
5. Điều kiện đặc biệt: 
- Nếu km > 30 và không phải ban đêm, áp dụng giảm 10% trên tổng tiền. 
- Nếu tổng tiền cuối cùng < 50,000 nhưng km > 5, đặt tổng tiền bằng 50,000 VND (mức tối thiểu).
khó vl cô ơi

```python
def taxi(lx, start_hours, km, cho):
    gt = 0
    if lx == 4:
        if km <= 0.8:
            gt = 11000
        elif km <= 20:
            gt = 11000 + (km - 0.8) * 13000
        else:
            gt = 11000 + (20 - 0.8) * 13000 + (km - 20) * 10500
    elif lx == 7:
        if km <= 0.8:
            gt = 12000
        elif km <= 20:
            gt = 12000 + (km - 0.8) * 14000
        else:
            gt = 12000 + (20 - 0.8) * 14000 + (km - 20) * 12000
    if start_hours in [22, 23, 0, 1, 2, 3, 4, 5]:
        gt *= 1.2
    if km > 30:
        gt *= 0.9
    if km > 5 and gt < 50000:
        gt = 50000
    tong = gt + (cho // 5) * 2000

    return tong
print(f"Tổng tiền: {taxi(7, 23, 45, 15):,.0f} VNĐ")
```

Viết hàm trả về góc ∠ABC theo đơn vị độ khi biết tọa độ các điểm A(x1,y1), B(x2,y2), C(x3,y3)

```python
import math
def goc(x1,y1,x2,y2,x3,y3):
    xa = x1 - x2
    ya = y1 - y2
    xc = x3 - x2
    yc = y3 - y2
    tvh = xa * xc + ya * yc
    tdai = (( xa** 2 + ya ** 2)** 0.5)*( (xc ** 2 + yc ** 2)** 0.5)
    cosx = tvh / tdai
    goc_rad = math.acos(cosx)        
    goc_deg = math.degrees(goc_rad)  
    return goc_deg
x1,y1,x2,y2,x3,y3 = map(int,input().split())
goc(x1,y1,x2,y2,x3,y3)
```
